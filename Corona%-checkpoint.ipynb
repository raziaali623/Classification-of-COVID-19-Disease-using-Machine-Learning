```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.preprocessing import StandardScaler

data = pd.read_excel("COVID-19 Clinical Data Repository.xlsx", 
                     sheet_name= ["04-07_carbonhealth_and_braidhea", 
                                  "04-14_carbonhealth_and_braidhea", 
                                  "04-21_carbonhealth_and_braidhea"])
sheets = pd.DataFrame()
sheets = data["04-07_carbonhealth_and_braidhea"]
sheets = sheets.append(data["04-14_carbonhealth_and_braidhea"])
sheets = sheets.append(data["04-21_carbonhealth_and_braidhea"])
sheets
```

```python
filtered_data = pd.DataFrame()
```

```python
# FILTER INCOMPLETE COLS
filtered_data["age"] = sheets["age"]
filtered_data["diabetes"] = sheets["diabetes"]
filtered_data["chd"] = sheets["chd"]
filtered_data["htn"] = sheets["htn"]
filtered_data["asthma"] = sheets["asthma"]
filtered_data["copd"] = sheets["copd"]
filtered_data["autoimmune_dis"] = sheets["autoimmune_dis"]
filtered_data["temperature"] = sheets["temperature"]
filtered_data["pulse"] = sheets["pulse"]
filtered_data["sys"] = sheets["sys"]
filtered_data["dia"] = sheets["dia"]
filtered_data["sats"] = sheets["sats"]
filtered_data["ctab"] = sheets["ctab"]
filtered_data["labored_respiration"] = sheets["labored_respiration"]
filtered_data["rhonchi"] = sheets["rhonchi"]
filtered_data["wheezes"] = sheets["wheezes"]
filtered_data["label"] = sheets["covid19_test_results"] # Test Result  
```

```python
filtered_data.dropna(subset = {"age", "diabetes", "chd", "htn", "asthma", "copd", "autoimmune_dis", "temperature", "pulse", "sys", "dia", "sats", "ctab", "labored_respiration", "rhonchi", "wheezes", "label"}, inplace=True)

##subset —If wespecify subset=[‘age’] then only the rows value as NaN will be removed.
##inplace — By default no changes will be made to your dataframe.
```

```python
filtered_data["label"].value_counts()
```

```python
filtered_data
```

```python
filtered_data = filtered_data.reset_index(drop=True)
filtered_data["asthma"] = filtered_data["asthma"].astype(int)
filtered_data["diabetes"] = filtered_data["diabetes"].astype(int)
filtered_data["chd"] = filtered_data["chd"].astype(int)
filtered_data["htn"] = filtered_data["htn"].astype(int)
filtered_data["copd"] = filtered_data["copd"].astype(int)
filtered_data["autoimmune_dis"] = filtered_data["autoimmune_dis"].astype(int)
filtered_data["ctab"] = filtered_data["ctab"].astype(int)
```

```python
positives = filtered_data.loc[filtered_data['label'] == "Positive"]
positives
```

```python
for i in range(3):
    positives = positives.append(positives)
positives.reset_index()
filtered_data = filtered_data.append(positives)
filtered_data
```

```python
labeless_data = filtered_data.drop("label", axis=1)
standardized_data = StandardScaler().fit_transform(labeless_data)
print (standardized_data)
## Convert numerical data 1 TO -1...so all columns have same range of numerical data.there are two reasons :algorithm uses euclidean distance.machine learning algos calculate distance between two points.if we not use this machine learning will not accurate as we want
```

```python
# PCA
from sklearn import decomposition
labels = filtered_data["label"]
pca = decomposition.PCA()
pca.n_components = 2
pca_data = pca.fit_transform(standardized_data)
print(pca_data.shape)
pca_data = np.vstack((pca_data.T, labels)).T
print(pca_data.shape)

pca_df = pd.DataFrame(data=pca_data, columns=('PCA_Component_1', 'PCA_Component_2', 'labels'))
print(pca_df)
```

```python
sns.FacetGrid(data=pca_df,hue='labels',height=6)\
    .map(plt.scatter, 'PCA_Component_1', 'PCA_Component_2')\
    .add_legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X = labeless_data
y = labels
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)


##The idea behind StandardScaler is that it will transform your data such that its distribution will have a mean value 0 and standard deviation of 1. In case of multivariate data, this is done feature-wise (in other words independently for each column of the data).
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()
clf=clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)


print('Accuracy of  classifier on test set: {:.2f}'
     .format(clf.score(X_test, y_test)))

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
pred = clf.predict(X_test)
print(confusion_matrix(y_test, pred))
print(classification_report(y_test, pred))


print('Accuracy of  classifier 10-Fold Cross-Validations Scores')

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, cv=10)
scores
```

```python
print('Accuracy of  classifier 5-Fold Cross-Validations Scores')

from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X, y, cv=5)
scores
```
